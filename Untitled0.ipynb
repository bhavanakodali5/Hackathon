```python
import pandas as pd
import random
import re
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report

# --- Setup ---
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# --- Generate synthetic video data ---
video_ids = [f"vid_{i:04d}" for i in range(1, 1001)]
durations = [round(random.uniform(5.0, 30.0), 1) for _ in range(1000)]
categories = ['comedy', 'sports', 'news', 'gaming', 'education', 'drama', 'prank', 'music']
labels = ['safe', 'unsafe']
notes_pool = [
    'funny cat video', 'intense fight scene', 'soccer goal moment', 'political debate',
    'jump scare prank', 'cooking tutorial', 'flash warning', 'blood and gore',
    'dance challenge', 'product promotion', 'alien invasion prank', 'news flash',
    'boring spreadsheet tutorial', 'subscribe for more', 'explosion clip',
    'music concert recording'
]

data = {
    'video_id': video_ids,
    'duration': durations,
    'category': [random.choice(categories) for _ in range(1000)],
    'label': [random.choice(labels) for _ in range(1000)],
    'notes': [random.choice(notes_pool) for _ in range(1000)]
}

df = pd.DataFrame(data)

# --- Text cleaning ---
df['notes_clean'] = df['notes'].apply(clean_text)
df['combined'] = df['category'] + ' ' + df['notes_clean']

# --- TF-IDF vectorization with bigrams ---
vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=3000)
X_text = vectorizer.fit_transform(df['combined'])

# --- Numerical feature ---
X_num = df[['duration']].values
X = np.hstack((X_text.toarray(), X_num))

# --- Labels ---
le = LabelEncoder()
y = le.fit_transform(df['label'])

# --- Train/test split with stratify ---
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X, y, df.index, test_size=0.2, random_state=42, stratify=y
)

# --- Random Forest model ---
model = RandomForestClassifier(
    n_estimators=500,
    max_depth=20,
    min_samples_split=5,
    class_weight='balanced',
    random_state=42
)
model.fit(X_train, y_train)

# --- Predict ---
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)

# --- Results ---
predictions = pd.DataFrame({
    'video_id': df.loc[idx_test, 'video_id'].values,
    'actual_label': le.inverse_transform(y_test),
    'predicted_label': le.inverse_transform(y_pred),
    'confidence_safe': y_prob[:, 0],
    'confidence_unsafe': y_prob[:, 1]
})

# --- Evaluation ---
print(predictions.head())
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```

```python
def tag_category(note):
    note = note.lower()
    if 'fight' in note or 'blood' in note or 'attack' in note:
        return 'violence'
    elif 'subscribe' in note or 'link below' in note or 'promo' in note:
        return 'spam'
    elif 'prank' in note or 'funny' in note:
        return 'comedy'
    else:
        return 'uncategorized'

df['tag'] = df['notes'].apply(tag_category)
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Load & clean data
df = pd.read_csv('VideoModLite.csv')
df['notes'] = df['notes'].fillna('')

# --- Text preprocessing ---
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text

df['notes_clean'] = df['notes'].apply(clean_text)

# --- Label encoding ---
le = LabelEncoder()
y = le.fit_transform(df['label'])

# --- Feature setup ---
tfidf = TfidfVectorizer(ngram_range=(1, 2), max_features=10000)
X_text = tfidf.fit_transform(df['notes_clean'])

# One-hot encode category
category_ohe = pd.get_dummies(df['category'])

# Duration
duration = df[['duration']].values

# Combine features
X = np.hstack((X_text.toarray(), category_ohe.values, duration))

# --- Train/Test split ---
X_train, X_test, y_train, y_test, train_idx, test_idx = train_test_split(
    X, y, df.index, test_size=0.2, random_state=42, stratify=y
)

# --- Model ---
model = LogisticRegression(max_iter=1000, class_weight='balanced', C=1.0)
model.fit(X_train, y_train)

# --- Predict ---
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

# --- Output ---
flagged_df = pd.DataFrame({
    'video_id': df.loc[test_idx, 'video_id'].values,
    'actual_label': le.inverse_transform(y_test),
    'predicted_label': le.inverse_transform(y_pred)
})
flagged_df['flagged'] = flagged_df['predicted_label'].apply(lambda x: 'Yes' if x == 'unsafe' else 'No')

print(f"Model Accuracy: {acc:.4f}")
print(flagged_df.head())

```
